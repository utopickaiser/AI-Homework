```python
# Correlation
import pandas as pd
import numpy as np
url = 'passenger.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
#Dollars (Y) phụ thuộc vào cái lớn nhất
from sklearn import linear_model
# Hồi quy tuyến tính bậc 1
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X = pd.DataFrame(data['Miles(X)']), y = data['Dollars (Y)']))
# pt hồi quy: y = coef*X + intercept
#data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black', xdim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['Miles(X)']))
plt.plot(data['Miles(X)'], y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2]).T
model.fit(X, y = data['Dollars (Y)'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 2', model.score(X, y = data['Dollars (Y)']))
range = np.arange(0, 5000, 100)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 3
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2 ,data['Miles(X)']**3]).T
model.fit(X, y = data['Dollars (Y)'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 3', model.score(X, y = data['Dollars (Y)']))
range = np.arange(0, 5000, 100)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 5
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3,
                  data['Miles(X)']**4, data['Miles(X)']**5]).T
model.fit(X, y = data['Dollars (Y)'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 5', model.score(X, y = data['Dollars (Y)']))
range = np.arange(0, 5000, 100)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 10
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3, 
                  data['Miles(X)']**4, data['Miles(X)']**5, data['Miles(X)']**6, 
                  data['Miles(X)']**7, data['Miles(X)']**8, data['Miles(X)']**9 ,data['Miles(X)']**10]).T
model.fit(X, y = data['Dollars (Y)'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 10', model.score(X, y = data['Dollars (Y)']))
range = np.arange(0, 5000, 100)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5, 
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 12
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3, 
                  data['Miles(X)']**4, data['Miles(X)']**5, data['Miles(X)']**6, 
                  data['Miles(X)']**7, data['Miles(X)']**8, data['Miles(X)']**9 ,data['Miles(X)']**10,
                 data['Miles(X)']**11, data['Miles(X)']**12]).T
model.fit(X, y = data['Dollars (Y)'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 12', model.score(X, y = data['Dollars (Y)']))
range = np.arange(0, 5000, 100)
data.plot(kind = 'scatter', x = 'Miles(X)', y = 'Dollars (Y)', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5, 
                          range**6, range**7, range**8, range**9, range**10,range**11, range**12]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')

```
