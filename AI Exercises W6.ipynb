```python
# Correlation
import pandas as pd
import numpy as np
url = 'mtcars.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
import seaborn as sns
sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
#mpg phụ thuộc vào cái lớn nhất
from sklearn import linear_model
# Hồi quy tuyến tính bậc 1
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác', model.score(X = pd.DataFrame(data['wt']), y = data['mpg']))
# pt hồi quy: y = coef*X + intercept
#data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black', xdim = (0,7))
y_pred = model.predict(X = pd.DataFrame(data['wt']))
plt.plot(data['wt'], y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 2', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 3
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2 ,data['wt']**3]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 3', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 5
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3,
                  data['wt']**4, data['wt']**5]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 5', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 10
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, 
                  data['wt']**4, data['wt']**5, data['wt']**6, 
                  data['wt']**7, data['wt']**8, data['wt']**9 ,data['wt']**10]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 10', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5, 
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
# Hồi quy tuyến tính bậc 12
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3, 
                  data['wt']**4, data['wt']**5, data['wt']**6, 
                  data['wt']**7, data['wt']**8, data['wt']**9 ,data['wt']**10,
                 data['wt']**11, data['wt']**12]).T
model.fit(X, y = data['mpg'])
print('Intercept', model.intercept_)
print('Coef', model.coef_)
print('Độ chính xác bậc 12', model.score(X, y = data['mpg']))
range = np.arange(1.5, 5.5, 0.1)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2, range**3, range**4, range**5, 
                          range**6, range**7, range**8, range**9, range**10,range**11, range**12]).T
y_pred = model.predict( X = predictor)
plt.plot( range, y_pred, color = 'blue')
```
